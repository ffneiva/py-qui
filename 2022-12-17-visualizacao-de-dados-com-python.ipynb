##### Importar bibliotecas utilizadas

```python
import pandas as pd
import plotly.express as px
```

##### Importar bases de dados

```python
df_election = px.data.election()
df_gapminder = px.data.gapminder()
df_iris = px.data.iris()
df_stocks = px.data.stocks(indexed=True)
df_tips = px.data.tips()
```

##### Paleta de cores

```python
cores = px.colors.qualitative # Pacote com várias paletas de cores
cores.swatches().show() # Mostrar paletas disponíveis 
```

```python
# Opções de paleta para gráficos Plotly
paleta = [cores.Set1, cores.Bold, cores.Dark2, cores.Safe, cores.D3, cores.G10,
          cores.T10, cores.Alphabet, cores.Prism, cores.Dark24, cores.Light24]
# Opções de fundo para os gráficos Plotly
fundo = ['plotly_dark', 'seaborn', 'simple_white', 'plotly', 
         'ggplot2', 'plotly_white', 'presentation', 
         'xgridoff', 'ygridoff', 'gridon', 'none']
# Opções de fontes para letras nos gráficos Plotly
fonte = ['Arial', 'Times New Roman', 'Montserrat', 'Opera Sans',
         'Lato', 'Roboto', 'Oswald', 'Source Sans Pro', 'Raleway',
         'PT Sans', 'Merriweather']
# Opções de cores para fontes de Plotly
cor_fonte = ['white', 'darkblue', 'green', 'brown', 'gold', 'crimson',
             'midnightblue', 'purple', 'cadetblue', 'deeppink', 'black']
```

### 2. Gráficos de distribuição

##### Histograma

```python
num = 0 # Número do tema
hist = px.histogram( # Método para fazer gráficos de histograma
    df_iris, # Dataframe utilizado no gráfico
    x='petal_length', # Coluna do eixo X
    nbins=100, # Número de intervalos
    color='species', # Coluna para separar por cores
    color_discrete_sequence=paleta[num], # Paleta de cores para utilizar 
    template=fundo[num] # Cor de fundo do gráfico
)
hist.update_layout( # Método para atualizar layout do gráfico
    title={ # Fazer o título do gráfico
        'text': 'Comprimento das pétalas de algumas flores', # Título do gráfico
        'font_family': fonte[num], # Seleciona a fonte utilizada no título
        'font_color': cor_fonte[num], # Seleciona a cor da fonte do título
        'font_size': 36, # Seleciona o tamanho do título
        'x': 0.5, # Centraliza o título
        'y': 0.95 # Abaixa um pouco o título (95% debaixo para cima da figura)
    }
)
hist.update_xaxes( # Método para atualizar o eixo X
    title={ # Editar o título do eixo X
        'text': 'Comprimento das pétalas', # Texto do título do eixo X
        'font_family': fonte[num], # Seleciona a fonte utilizada no eixo X
        'font_color': cor_fonte[num], # Altera a cor da fonte no eixo X
        'font_size': 20 # Altera o tamanho do texto no eixo X
    },
    color=cor_fonte[num] # Alterar a cor dos valores do eixo X
)
hist.update_yaxes( # Método para atualizar o eixo Y
    title={ # Editar o título do eixo Y
        'text': 'Quantidade', # Texto do título do eixo Y
        'font_family': fonte[num], # Seleciona a fonte utilizada no eixo Y
        'font_color': cor_fonte[num], # Altera a cor da fonte no eixo Y
        'font_size': 20 # Altera o tamanho do texto no eixo Y
    },
    color=cor_fonte[num] # Alterar a cor dos valores do eixo Y
)
hist # Mostrar gráfico
```

### 3. Gráficos de correlação

##### Gráfico de dispersão

```python
num = 3 # Número do tema
dispersao = px.scatter( # Método utilizado para fazer o gráfico de dispersão
    df_tips, # Dataframe utilizado no gráfico
    x='total_bill', # Coluna para o eixo Y
    y='tip', # Coluna para o eixo Y
    color='sex', # Coluna para separar por cores
    size='size', # Tamanho dos pontos do gráfico
    # symbol='sex', # Separar as informações por símbolos
    trendline='ols', # Regressão linear
    trendline_scope='overall', # Para todos os dados (sem separar por cor)
    trendline_color_override='black', # Cor da curva da regressão
    color_discrete_sequence=paleta[num], # Paleta de cores para utilizar 
    template=fundo[num] # Cor de fundo do gráfico
)
dispersao.update_layout(
    title={ # Fazer o título do gráfico
        'text': 'Gorjeta versus valor da conta', # Título do gráfico
        'font_family': fonte[num], # Seleciona a fonte utilizada no título
        'font_color': cor_fonte[num], # Seleciona a cor da fonte do título
        'font_size': 36, # Seleciona o tamanho do título
        'x': 0.5, # Centraliza o título
        'y': 0.95 # Abaixa um pouco o título (95% debaixo para cima da figura)
    }
)
dispersao.update_xaxes(
    title={ # Editar o título do eixo X
        'text': 'Valor da conta', # Texto do título do eixo X
        'font_family': fonte[num], # Seleciona a fonte utilizada no eixo X
        'font_color': cor_fonte[num], # Altera a cor da fonte no eixo X
        'font_size': 20 # Altera o tamanho do texto no eixo X
    },
    color=cor_fonte[num] # Alterar a cor dos valores do eixo X
)
dispersao.update_yaxes(
    title={ # Editar o título do eixo Y
        'text': 'Gorjeta', # Texto do título do eixo Y
        'font_family': fonte[num], # Seleciona a fonte utilizada no eixo Y
        'font_color': cor_fonte[num], # Altera a cor da fonte no eixo Y
        'font_size': 20 # Altera o tamanho do texto no eixo Y
    },
    color=cor_fonte[num] # Alterar a cor dos valores do eixo Y
)
dispersao # Mostrar gráfico
```

##### Gráfico de radar

```python
num = 7 # Número do tema
df_ano = df_gapminder.query('year == 2007') # Selecionar somente ano especificado
df_media = df_ano.groupby('continent').mean() # Agrupar fazendo a média
radar = px.line_polar( # Método para fazer gráficos de radar
    df_media, # Dataframe utilizado no gráfico
    r='lifeExp', # Valor dos raios
    theta=df_media.index, # Legenda dos valores (ângulos)
    text='lifeExp', # Mostrar rótulo de dados
    line_close=True, # Unir linhas do gráfico
    color_discrete_sequence=paleta[num], # Paleta de cores para utilizar 
    template=fundo[num] # Cor de fundo do gráfico
)
radar.update_traces(fill='toself') # Colorir área do gráfico
radar.update_layout( # Método para atualizar layout do gráfico
    title={ # Fazer o título do gráfico
        'text': 'Expectativa de vida média em cada continente em 2017', # Título do gráfico
        'font_family': fonte[num], # Seleciona a fonte utilizada no título
        'font_color': cor_fonte[num], # Seleciona a cor da fonte do título
        'font_size': 28, # Seleciona o tamanho do título
        'x': 0.5, # Centraliza o título
    }
)
radar # Mostrar gráfico
```

### Bônus 1: Nuvem de palavras

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_gradient_magnitude
from wordcloud import WordCloud, ImageColorGenerator

def criar_nuvem_de_palavras(texto, imagem, dividir=3):
    # Abre o arquivo de texto
    text = open(texto, encoding='utf-8').read()
    
    # Abre a imagem e carrega as cores dela
    colors = np.array(Image.open(imagem))
    # Diminuindo a quantidade de pixels para não demorar muito ao gerar a nuvem de palavras
    colors = colors[::dividir, ::dividir]

    # Criar uma máscara branca
    mask = colors.copy()
    mask[mask.sum(axis=2) == 0] = 255

    # Litando as cores para que elas fiquem menos desbotadas
    edges = np.mean([gaussian_gradient_magnitude(colors[:, :, i] / 255., 2) for i in range(3)], axis=0)
    mask[edges > .08] = 255

    # Criação da nuvem de palavras
    wc = WordCloud(max_words=2000, mask=mask, max_font_size=40, random_state=42, relative_scaling=0)
    wc.generate(text)

    # Criar coloração a partir da imagem
    image_colors = ImageColorGenerator(colors)
    return wc.recolor(color_func=image_colors)
```

```python
texto_papagaio = '/content/drive/MyDrive/0. Pasta/Exemplos/papagaio.txt'
imagem_papagaio = '/content/drive/MyDrive/0. Pasta/Exemplos/papagaio.jpg'
fig_papagaio = criar_nuvem_de_palavras(texto_papagaio, imagem_papagaio, dividir=5)
plt.axis('off')
plt.imshow(fig_papagaio)
```
